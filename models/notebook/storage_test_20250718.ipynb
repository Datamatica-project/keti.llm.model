```python
import boto3 

# S3 클라이언트 생성
s3 = boto3.client(
    's3',
    endpoint_url='http://artifact-store:9000',  # 또는 'http://artifact-store:9000'
    aws_access_key_id='minio',
    aws_secret_access_key='miniostorage',
)

# 대상 버킷
bucket_name = 'chunk'

# 리스트 가져오기
response = s3.list_objects_v2(Bucket=bucket_name, Prefix="data/")

if 'Contents' in response:
    files = [obj['Key'] for obj in response['Contents']]
    print("📄 파일 리스트:")
    for file in files:
        print(file)
else:
    print("❗버킷이 비어있거나 해당 Prefix에 파일이 없습니다.")
```

```python
import boto3
import json

# MinIO 클라이언트 설정
s3_client = boto3.client(
    "s3",
    endpoint_url="http://artifact-store:9000",  # 또는 MinIO 주소
    aws_access_key_id="minio",
    aws_secret_access_key="miniostorage"
)

bucket_name = "chunk"
key = "data/30 밀_저품질_단면_2025-07-18_cc058052-4.json"  # 가져올 JSON 파일 경로

# JSON 파일 가져오기
response = s3_client.get_object(Bucket=bucket_name, Key=key)
content = response['Body'].read().decode('utf-8')
data = json.loads(content)

# 이제 data는 Python 변수입니다
print(type(data))  # dict 또는 list
data
```

```python
for lines in data:
    lines["document"] = "30 밀_저품질_단면.pdf"
```

```python
data
```

```python
import io 

buffer = io.BytesIO()
buffer.write(json.dumps(data, ensure_ascii=False).encode("utf-8"))
buffer.seek(0)

s3_client.upload_fileobj(buffer, bucket_name, key)
```

```python
from datasets import Dataset
import boto3
from io import BytesIO
import pandas as pd

# boto3 클라이언트 설정
client = boto3.client(
    's3',
    endpoint_url="http://artifact-store:9000",
    aws_access_key_id="minio",
    aws_secret_access_key="miniostorage"
)

# 방법 1: pandas 경유 (추천)
response = client.get_object(Bucket="instruction", Key="data/0.0.1v/agriculture.trainset.parquet")
parquet_data = response['Body'].read()

# pandas로 읽고 Dataset으로 변환
df = pd.read_parquet(BytesIO(parquet_data))
dataset = Dataset.from_pandas(df)

print(f"데이터 크기: {len(dataset)}")
print(f"컬럼: {dataset.column_names}")
```
